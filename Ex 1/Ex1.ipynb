```python
# (i)

import numpy as np

dataa = np.loadtxt("data_1a.dat")
datab = np.loadtxt("data_1b.dat")
datac = np.loadtxt("data_1c.dat")
datad = np.loadtxt("data_1d.dat")
```

```python
# (ii)

import seaborn as sns
from matplotlib import pyplot as plt

sns.set()

fig, ax = plt.subplots(2,2, figsize=(10, 6), layout='constrained')

ax[0,0].set_title("Data A")
ax[0,1].set_title("Data B")
ax[1,0].set_title("Data C")
ax[1,1].set_title("Data D")

ax[0,0].set_ylabel("Frequency of x")
ax[0,1].set_ylabel("Frequency of x")
ax[1,0].set_ylabel("Frequency of x")
ax[1,1].set_ylabel("Frequency of x")
ax[0,0].set_xlabel('Variable x')
ax[0,1].set_xlabel('Variable x')
ax[1,0].set_xlabel('Variable x')
ax[1,1].set_xlabel('Variable x')

ax[0,0].hist(dataa, 25, edgecolor='none')
ax[0,1].hist(datab, 14, edgecolor='none')
ax[1,0].hist(datac, 30, edgecolor='none')
ax[1,1].hist(datad, 50, edgecolor='none')
plt.show()
```

```python

#https://matplotlib.org/stable/api/legend_api.html
import matplotlib.patches as mpatches
#colorcycle 
sns.set_style("whitegrid")
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

a_patch = mpatches.Patch(color=cycle[0], label='data_1a.dat', alpha = 0.8)
b_patch = mpatches.Patch(color=cycle[1], label='data_1b.dat', alpha = 0.5)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6), layout='constrained')


#https://numpy.org/doc/stable/reference/generated/numpy.histogram.html
histoa = np.histogram(dataa, bins = 1000);
histob = np.histogram(datab, bins = 1000);
ax1.bar(histoa[1][0:len(histoa[1])-1], histoa[0], alpha = 0.8)
ax1.bar(histob[1][0:len(histob[1])-1], histob[0], alpha = 0.5)
ax1.set_ylabel('Frequency of x', color = 'black', fontsize='12')
ax1.set_xlabel('Variable x', color = 'black', fontsize='12')
ax1.set_title("Data A and B", fontsize='12')
ax1.legend(handles=[a_patch, b_patch])

#1c and 1d#
ax2.set_title("Data C and D", fontsize='12')

c_patch = mpatches.Patch(color=cycle[0], label='data_1c.dat', alpha = 1)
d_patch = mpatches.Patch(color=cycle[1], label='data_1d.dat', alpha = 0.5)


#https://de.mathworks.com/help/matlab/ref/matlab.graphics.chart.primitive.bar-properties.html
ax2.hist(datac, 30, alpha = 1, edgecolor = 'none')#, color = 'orange', edgecolor = "darkorange", linewidth = 0 , mouseover = 'true')
ax2.hist(datad, 50, alpha = 0.5, edgecolor = 'none' )#, color = 'orange', edgecolor = "darkorange", linewidth = 0 , mouseover = 'true')
ax2.legend(handles=[c_patch, d_patch])
ax2.set_ylabel('Frequency of x', color = 'black', fontsize='12')
ax2.set_xlabel('Variable x', color = 'black', fontsize='12')
ax2.set_yscale("log")

plt.show()
```

### Excercise 2: Using prior knowledge
#### a) (1 Point):
Suppose there is a town with green and yellow taxicabs. In a hit-and-run accident a man was hurt and a witness reports to have seen a green taxi cab.
In court the lawyer of the taxi company impeaches the credibility of the witness, because of the lighting conditions. A test shows that under similar lighting conditions 10 \% report the respective wrong color when  seeing a taxi cab of either color.
What is your initial judgment for how likely it is for an honest witness to have correctly reported a green cab (this is not a trick question (yet))? 

Would you believe the witness?


$\color{blue}{\text{Since 10$\, \%$ report the respective color wrong, the likeliness of the witness reporting the correct color is 90$\,\%$.}}$

#### b) (4 Points):
Suppose the town has 200 Yellow taxi cabs and only 10 green taxi cabs. Using this prior knowledge, *(i)* calculate the probability of having actually seen a yellow cab after reporting a green cab under the previously described lighting conditions as well as *(ii)* the probability of actually having seen a green cab after reporting one. 

Would you believe the Witness now?

<span style="color:blue">

(i) We want to calculate P(yellow | green r), where only the color marks the real color of the cab and an additional "r" marks the reported color. The equation from the lecture reads $\newline$
$P(yellow| green\, r)=\frac{P(green\, r| yellow) \cdot P(yellow)}{P(green\, r)}$. $\newline$


We already know that $P(green\, r| yellow)=0.1$ and can calculate $P(yellow)$ from the number of yellow cars (200) divided by the total number of cars (210). So $P(yellow)=\frac{20}{21}$. $\newline$
Then $P(green\,r)=P(green\, r|green)\cdot P(green)+P(green\,r|yellow)\cdot P(yellow)=0.9 \cdot \frac{1}{21} + 0.1 \cdot \frac{20}{21} = \frac{29}{210} \approx 13.8 \, \%$. $\newline$
With this we get the final result $P(yellow| green\, r)= \frac{0.1 \cdot \frac{20}{21}}{\frac{29}{210}} \approx 69.0 \, \%$.

So the probability is around 69%.

(ii) Using the same formula we now want to calculate $P(green| green\, r)=\frac{P(green\, r|green) \cdot P(green)}{P(green\, r)}$. $\newline$
$P(green\, r|green)=0.9$, $P(green)=\frac{1}{21}$ and $P(green\,r)=0.14$ as before. Then

$P(green| green\, r)=0.31$,

so the probability is around $31\,\%$. $\newline$
As expected the two probabilities add up to $100 \, \%$.

The chances that the witness actually saw a green cab are around half the chance of wrongfully reporting a green one. But if the witness saw it correctly it would be far easier to catch the felon among the 10 cabs! </span>
